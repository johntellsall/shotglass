```python


from matplotlib import pyplot as plt

sns.set_theme(context='talk', palette='rainbow')
plot_container, axs = plt.subplots(1, 2, figsize=(8, 4), gridspec_kw=dict(width_ratios=[4, 3]))
sns.scatterplot(data=penguins, hue="species", x="flipper_length_mm", y="bill_length_mm", ax=axs[0])
sns.histplot(data=penguins, x="species", hue="species", shrink=0.8, alpha=.6, legend=False, ax=axs[1])
plot_container.tight_layout()
```

```python
tips = sns.load_dataset("tips")
g = sns.relplot(data=tips, x="total_bill", y="tip")
g.ax.axline(xy1=(10, 2), slope=.2, color="b", dashes=(5,2))


```
